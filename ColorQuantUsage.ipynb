```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ColorQuantization import ColorQuantization, load_image, count_colors
%matplotlib inline

plt.style.use('ggplot')
```

```python
# Load Image
FILE = 'china.png'

img = load_image(FILE)
plt.axis('off')
plt.imshow(img)
```

```python
# Declare CQ KMEANS and plot pixels
cq = ColorQuantization(n_colors=8)

cq.plot_pixels(img, title=f"Input color space: {count_colors(img)} unique colors")
```

```python
# KMEANS
cq.fit([img])
new_img = cq.colorize([img])[0]
cq.plot_pixels(img, title="Reduced color space: 8 colors", colors=new_img)
```

```python
# display both imgs
cq.plot_img(img, new_img)
```

```python
# GMM
cq_gmm = ColorQuantization(n_colors=8, algo='gmm')
cq_gmm.fit([img])
new_img = cq_gmm.colorize([img])[0]
cq_gmm.plot_pixels(img, title="Reduced color space: 8 colors", colors=new_img)
```

```python
# display both imgs
cq_gmm.plot_img(img, new_img)
```
