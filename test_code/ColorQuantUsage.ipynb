```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from ColorQuantization import ColorQuantization, load_image, count_colors
from sklearn.datasets import load_sample_image

%matplotlib inline
```

```python
# Load Image

# img = load_sample_image('china.jpg') 
img = load_sample_image('flower.jpg') 
plt.imshow(img)
```

```python
# Declare CQ KMEANS and plot pixels
cq = ColorQuantization(n_colors=4)
cq.fit([img])
new_img = cq.colorize([img])[0]
print(img.shape)
```

```python
cq.plot_pixels(img, title=f"Input color space: {count_colors(img)} unique colors")
```

```python
cq.plot_pixels(img, title=f"Reduced color space: {cq.n_colors} colors", colors=new_img)
```

```python
cq.plot_img(img, new_img)
```

```python
# GMM
cq_gmm = ColorQuantization(n_colors=8, algo='gmm')
cq_gmm.fit([img])
new_img = cq_gmm.colorize([img])[0]
cq_gmm.plot_pixels(img, title="Reduced color space: 8 colors", colors=new_img)
```

```python
# display both imgs
cq_gmm.plot_img(img, new_img)
```
